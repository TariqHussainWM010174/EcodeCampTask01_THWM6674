First we are going to change the SeniourCitizen to move 'yes' and 'no' by creating a function and then replacing the values
in the dataframe

```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
# Replacing the TotalCharges value as Zero due to that the customer did not start any plan with the company
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")   # Replacing the TotalCharges values as 0
df["TotalCharges"] = df["TotalCharges"].astype("float")    # And Totalcharges column should be numerical column so we are changing the datatype
```

Features Relationships (Bivariate Analysis)
We are gonig to do the Bivariate Analysis, by taking the features (categorical, numerical) with the traget (Churn)
Categorical Features with Target variable (Churn)

```python
plt.figure(figsize = (4,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
#columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid (you can change this)
n_cols = 3
n_rows = (len(categorical_features) + n_cols - 1) // n_cols  # Calculate number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))  # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays)
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(categorical_features):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'{col} wise Churn')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (6,4))
ax = sns.countplot(x = "PaymentMethod", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation = 45)
plt.show()
```
